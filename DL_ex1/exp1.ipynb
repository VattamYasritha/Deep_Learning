```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, FancyBboxPatch

# Create figure
fig, ax = plt.subplots(figsize=(16,7))
ax.set_xlim(0,18)
ax.set_ylim(0,8)
ax.axis('off')

# ===========================
# SINGLE LAYER PERCEPTRON
# ===========================

# Input Neurons
single_inputs = [(1.5,5.5),(1.5,2.5)]

for x,y in single_inputs:
    ax.add_patch(Circle((x,y),0.28,
                        facecolor="#9B59B6",
                        edgecolor="black",
                        linewidth=2))

# Labels
ax.text(1.5,5.5,"x₁",ha='center',va='center',color='white',fontweight='bold')
ax.text(1.5,2.5,"x₂",ha='center',va='center',color='white',fontweight='bold')

# Perceptron Block
box = FancyBboxPatch((4,3.4),2.5,1.4,
                     boxstyle="round,pad=0.2",
                     facecolor="#48C9B0",
                     edgecolor="black",
                     linewidth=2)

ax.add_patch(box)

ax.text(5.25,4.1,
        "Weighted Sum\n+ Sigmoid",
        ha='center',
        va='center',
        fontsize=10,
        fontweight='bold')

# Output
ax.add_patch(Circle((8.5,4.1),0.3,
                    facecolor="#F39C12",
                    edgecolor="black",
                    linewidth=2))

ax.text(8.5,4.1,"y",
        ha='center',
        va='center',
        fontweight='bold')

# Connections
for x,y in single_inputs:
    ax.annotate("",xy=(4,y+0.2),xytext=(1.8,y),
                arrowprops=dict(arrowstyle="->",lw=2))

ax.annotate("",xy=(8.2,4.1),xytext=(6.5,4.1),
            arrowprops=dict(arrowstyle="->",lw=2))

# Bias
ax.annotate("",xy=(5.25,3.4),xytext=(5.25,1.2),
            arrowprops=dict(arrowstyle="->",lw=2,color="red"))

ax.text(5.25,0.8,"Bias",
        ha='center',
        color="red",
        fontweight='bold')

ax.text(4.5,6.8,
        "Single-Layer Perceptron",
        fontsize=14,
        fontweight='bold',
        color="#2E4053",
        ha='center')

# ===========================
# MULTI LAYER PERCEPTRON
# ===========================

# Input Layer
inputs=[(11,5.5),(11,2.5)]

# Hidden Layer
hidden=[(13.8,6),(13.8,4),(13.8,2)]

# Output Layer
output=(16.5,4)

# Draw Input
for x,y in inputs:
    ax.add_patch(Circle((x,y),0.28,
                        facecolor="#6C5CE7",
                        edgecolor="black",
                        linewidth=2))

# Draw Hidden
for x,y in hidden:
    ax.add_patch(Circle((x,y),0.28,
                        facecolor="#2ECC71",
                        edgecolor="black",
                        linewidth=2))

# Draw Output
ax.add_patch(Circle(output,0.3,
                    facecolor="#E67E22",
                    edgecolor="black",
                    linewidth=2))

# Connections
for ix,iy in inputs:
    for hx,hy in hidden:
        ax.plot([ix,hx],[iy,hy],
                color="gray",
                linewidth=2)

for hx,hy in hidden:
    ax.plot([hx,output[0]],[hy,output[1]],
            color="gray",
            linewidth=2)

# Labels
ax.text(11,5.5,"x₁",
        ha='center',
        va='center',
        color='white',
        fontweight='bold')

ax.text(11,2.5,"x₂",
        ha='center',
        va='center',
        color='white',
        fontweight='bold')

ax.text(14.2,6,"H₁")
ax.text(14.2,4,"H₂")
ax.text(14.2,2,"H₃")

ax.text(16.5,4,"y",
        ha='center',
        va='center',
        fontweight='bold')

# Layer Names
ax.text(11,6.8,"Input",
        fontsize=12,
        fontweight='bold')

ax.text(13.8,6.8,"Hidden",
        fontsize=12,
        fontweight='bold',
        ha='center')

ax.text(16.5,6.8,"Output",
        fontsize=12,
        fontweight='bold',
        ha='center')

ax.text(13.8,7.5,
        "Multi-Layer Perceptron",
        fontsize=14,
        fontweight='bold',
        color="#2E4053",
        ha='center')

# Overall Title
plt.suptitle("Comparison of Single-Layer Perceptron and Multi-Layer Perceptron",
             fontsize=18,
             fontweight='bold',
             color="#1B2631")

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Create Figure
fig, ax = plt.subplots(figsize=(11,6))
ax.set_xlim(0,12)
ax.set_ylim(0,10)
ax.axis('off')

# -----------------------------
# Neuron Positions
# -----------------------------

# Input Layer
inputs = [(1.5,7), (1.5,3)]

# Hidden Layer
hidden = [(6,8.5),
          (6,6.8),
          (6,5),
          (6,3.2),
          (6,1.5)]

# Output Layer
output = (10.5,5)

# -----------------------------
# Connections
# -----------------------------

for ix, iy in inputs:
    for hx, hy in hidden:
        ax.plot([ix, hx], [iy, hy],
                color="#5D6D7E",
                linewidth=1.6)

for hx, hy in hidden:
    ax.plot([hx, output[0]], [hy, output[1]],
            color="#34495E",
            linewidth=1.8)

# -----------------------------
# Draw Input Neurons
# -----------------------------

for x, y in inputs:
    ax.add_patch(
        Circle((x,y),0.28,
               facecolor="#7B68EE",
               edgecolor="black",
               linewidth=2)
    )

# -----------------------------
# Draw Hidden Neurons
# -----------------------------

for i,(x,y) in enumerate(hidden):
    ax.add_patch(
        Circle((x,y),0.30,
               facecolor="#58D68D",
               edgecolor="black",
               linewidth=2)
    )

    ax.text(x,y,
            f"H{i+1}",
            ha='center',
            va='center',
            fontsize=10,
            fontweight='bold')

# -----------------------------
# Draw Output Neuron
# -----------------------------

ax.add_patch(
    Circle(output,0.34,
           facecolor="#F5B041",
           edgecolor="black",
           linewidth=2)
)

ax.text(output[0],output[1],
        "y",
        ha='center',
        va='center',
        fontsize=12,
        fontweight='bold')

# -----------------------------
# Input Labels
# -----------------------------

ax.text(1.5,7,"x₁",
        color="white",
        ha='center',
        va='center',
        fontsize=11,
        fontweight='bold')

ax.text(1.5,3,"x₂",
        color="white",
        ha='center',
        va='center',
        fontsize=11,
        fontweight='bold')

# -----------------------------
# Layer Labels
# -----------------------------

ax.text(1.5,9.4,
        "Input Layer",
        fontsize=13,
        fontweight='bold',
        color="#4A235A",
        ha='center')

ax.text(6,9.6,
        "Hidden Layer\n(ReLU Activation)",
        fontsize=13,
        fontweight='bold',
        color="#145A32",
        ha='center')

ax.text(10.5,7.2,
        "Output Layer\n(Sigmoid)",
        fontsize=13,
        fontweight='bold',
        color="#7D6608",
        ha='center')

# -----------------------------
# Title
# -----------------------------

plt.title(
    "Multi-Layer Perceptron (MLP) Architecture for XOR Classification",
    fontsize=17,
    fontweight='bold',
    color="#1B2631",
    pad=20
)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# XOR Dataset
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,0])

plt.figure(figsize=(6,6))

for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i,0], X[i,1],
                    color="#FF6B6B",
                    marker="o",
                    s=250,
                    edgecolor="black",
                    label="Class 0" if i==0 else "")
    else:
        plt.scatter(X[i,0], X[i,1],
                    color="#4ECDC4",
                    marker="D",
                    s=250,
                    edgecolor="black",
                    label="Class 1" if i==1 else "")

for i in range(len(X)):
    plt.text(X[i,0]+0.03,
             X[i,1]+0.03,
             str(tuple(X[i])),
             fontsize=10)

plt.title("2D Visualization of XOR Dataset",fontsize=15,fontweight='bold')
plt.xlabel("Input x1")
plt.ylabel("Input x2")
plt.grid(True,linestyle="--",alpha=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf

# Step 1: Define XOR Dataset
X = np.array([[0,0], [0,1], [1,0], [1,1]], dtype=np.float32)
y = np.array([[0], [1], [1], [0]], dtype=np.float32)

# Step 2: Build the Multi-Layer Perceptron Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(5, input_dim=2, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Step 3: Compile Model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Step 4: Train Model
history = model.fit(X, y, epochs=3000, verbose=0)

# Step 5: Generate Predictions
raw_preds = model.predict(X, verbose=0)
class_preds = (raw_preds > 0.5).astype(int)

# Step 6: Formatted Output
print("-" * 45)
print(f"{'x1':<6} {'x2':<6} {'Target':<10} {'Prob':<10} {'Predicted':<10}")
print("-" * 45)
for i in range(len(X)):
    print(f"{int(X[i][0]):<6} {int(X[i][1]):<6} {int(y[i][0]):<10} {raw_preds[i][0]:<10.4f} {class_preds[i][0]:<10}")
print("-" * 45)

loss, accuracy = model.evaluate(X, y, verbose=0)
print(f"\nFinal Loss: {loss:.4f}")
print(f"Training Accuracy: {accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(history.history['loss'], color='#2E86C1', linewidth=2, label='Training Loss')

plt.title('Loss Convergence Curve', fontsize=14, fontweight='bold')
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Binary Cross-Entropy Loss', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix

# Step 1: Summary Table Generation
df = pd.DataFrame({
    'Input x1': X[:, 0].astype(int),
    'Input x2': X[:, 1].astype(int),
    'Actual Label': y.flatten().astype(int),
    'Predicted Probability': np.round(raw_preds.flatten(), 4),
    'Predicted Label': class_preds.flatten()
})

print("--- SUMMARY EVALUATION TABLE ---")
print(df.to_string(index=False))

# Step 2: Confusion Matrix
cm = confusion_matrix(y, class_preds)

print("\n--- CONFUSION MATRIX ---")
print(f"TN: {cm[0][0]}  FP: {cm[0][1]}")
print(f"FN: {cm[1][0]}  TP: {cm[1][1]}")

print(f"\nFinal Prediction Accuracy: {accuracy * 100:.2f}%")
```
