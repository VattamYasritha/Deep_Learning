```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(output):
    return output * (1 - output)
```

```python
# Input
x = np.array([[1, 0]])

# Target
y = np.array([[1]])

# Hidden layer weights
W1 = np.array([
    [0.2, 0.3],
    [0.4, 0.1]
])

# Output layer weights
W2 = np.array([
    [0.5],
    [0.6]
])

# Biases
b1 = np.array([[0.0, 0.0]])
b2 = np.array([[0.0]])

# Learning rate
learning_rate = 0.1
```

```python
# Hidden layer
hidden_input = np.dot(x, W1) + b1
hidden_output = sigmoid(hidden_input)

# Output layer
final_input = np.dot(hidden_output, W2) + b2
prediction = sigmoid(final_input)

print("Hidden Layer Output")
print(hidden_output)

print("\nPredicted Output")
print(prediction)
```

```python
loss = np.mean((y - prediction) ** 2)

print("\nMean Squared Error")
print(loss)
```

```python
# Output layer error
output_error = prediction - y

# Gradient at output
d_output = output_error * sigmoid_derivative(prediction)

# Gradient for W2
dW2 = np.dot(hidden_output.T, d_output)

# Hidden layer error
hidden_error = np.dot(d_output, W2.T)

# Hidden layer gradient
d_hidden = hidden_error * sigmoid_derivative(hidden_output)

# Gradient for W1
dW1 = np.dot(x.T, d_hidden)
```

```python
print("Gradient of W1")
print(dW1)

print("\nGradient of W2")
print(dW2)
```

```python
W1 = W1 - learning_rate * dW1

W2 = W2 - learning_rate * dW2
```

```python
b1 = b1 - learning_rate * d_hidden

b2 = b2 - learning_rate * d_output
```

```python
print("Updated W1")
print(W1)

print("\nUpdated W2")
print(W2)

print("\nUpdated b1")
print(b1)

print("\nUpdated b2")
print(b2)
```

```python
hidden_input = np.dot(x, W1) + b1
hidden_output = sigmoid(hidden_input)

final_input = np.dot(hidden_output, W2) + b2
prediction = sigmoid(final_input)

new_loss = np.mean((y - prediction) ** 2)

print("Prediction After Weight Update")
print(prediction)

print("\nNew Loss")
print(new_loss)
```

```python
print("Old Loss :", loss)
print("New Loss :", new_loss)

if new_loss < loss:
    print("\nTraining Successful")
else:
    print("\nLoss Did Not Reduce")
```
