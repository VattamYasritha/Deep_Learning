```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
ax = plt.gca()
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.axis('off')

# Coordinates of neurons
input_layer = [(0.15,0.75),(0.15,0.25)]
hidden_layer = [(0.50,0.75),(0.50,0.25)]
output_layer = [(0.85,0.50)]

# Draw neurons
for layer in [input_layer, hidden_layer, output_layer]:
    for x,y in layer:
        circle = plt.Circle((x,y),0.04,
                            fill=False,
                            linewidth=2)
        ax.add_patch(circle)

# Connections
for i in input_layer:
    for h in hidden_layer:
        plt.plot([i[0],h[0]],[i[1],h[1]],linewidth=1.8)

for h in hidden_layer:
    plt.plot([h[0],output_layer[0][0]],
             [h[1],output_layer[0][1]],
             linewidth=1.8)

# Labels
plt.text(0.12,0.80,"x₁",fontsize=12)
plt.text(0.12,0.30,"x₂",fontsize=12)

plt.text(0.47,0.80,"h₁",fontsize=12)
plt.text(0.47,0.30,"h₂",fontsize=12)

plt.text(0.90,0.52,"ŷ",fontsize=12)

plt.title("2-2-1 Neural Network")
plt.show()
```

```python
import numpy as np

def sigmoid(z):
    return 1/(1+np.exp(-z))

W1 = np.array([[0.2,0.3],
               [0.4,0.1]])

W2 = np.array([[0.5],
               [0.6]])

x = np.array([[1,0]])

# Forward Pass
hidden_input = x @ W1
hidden_output = sigmoid(hidden_input)

output_input = hidden_output @ W2
prediction = sigmoid(output_input)

print("="*45)
print(" FORWARD PROPAGATION ")
print("="*45)

print("\nHidden Layer Net Input")
print(np.round(hidden_input,4))

print("\nHidden Layer Output")
print(np.round(hidden_output,4))

print("\nOutput Layer Net Input")
print(np.round(output_input,4))

print("\nNetwork Prediction")
print(np.round(prediction,4))
```

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Given values
W1 = np.array([[0.2, 0.3],
               [0.4, 0.1]])

W2 = np.array([[0.5],
               [0.6]])

x = np.array([[1, 0]])
target = np.array([[1]])

# Forward Pass
hidden = sigmoid(np.dot(x, W1))
prediction = sigmoid(np.dot(hidden, W2))

# Mean Squared Error
mse_loss = 0.5 * np.square(target - prediction)

print("="*45)
print(" LOSS CALCULATION ")
print("="*45)

print("\nPredicted Output")
print(np.round(prediction,4))

print("\nMean Squared Error")
print(np.round(mse_loss,4))
```

```python
import numpy as np

def sigmoid(z):
    return 1/(1+np.exp(-z))

W1 = np.array([[0.2,0.3],
               [0.4,0.1]])

W2 = np.array([[0.5],
               [0.6]])

x = np.array([[1,0]])
y = np.array([[1]])

# Forward Pass
hidden = sigmoid(x @ W1)
prediction = sigmoid(hidden @ W2)

# Backpropagation
output_delta = (prediction - y) * prediction * (1 - prediction)

gradient_W2 = hidden.T @ output_delta

hidden_delta = (output_delta @ W2.T) * hidden * (1-hidden)

gradient_W1 = x.T @ hidden_delta

print("="*45)
print(" BACKPROPAGATION ")
print("="*45)

print("\nGradient of W2")
print(np.round(gradient_W2,6))

print("\nGradient of W1")
print(np.round(gradient_W1,6))
```

```python
import numpy as np

def sigmoid(z):
    return 1/(1+np.exp(-z))

# Initial Data
W1 = np.array([[0.2,0.3],
               [0.4,0.1]],dtype=float)

W2 = np.array([[0.5],
               [0.6]],dtype=float)

X = np.array([[1,0]],dtype=float)
Y = np.array([[1]],dtype=float)

lr = 0.1

# ---------- Forward Pass ----------
hidden_input = np.dot(X,W1)
hidden_output = sigmoid(hidden_input)

output_input = np.dot(hidden_output,W2)
prediction = sigmoid(output_input)

loss = 0.5*np.square(Y-prediction)

# ---------- Backpropagation ----------
delta_out = (prediction-Y) * prediction * (1-prediction)

gradW2 = np.dot(hidden_output.T,delta_out)

delta_hidden = np.dot(delta_out,W2.T) * hidden_output * (1-hidden_output)

gradW1 = np.dot(X.T,delta_hidden)

# ---------- Weight Update ----------
new_W1 = W1 - lr*gradW1
new_W2 = W2 - lr*gradW2

# ---------- Display ----------
print("="*55)
print("BACKPROPAGATION VERIFICATION")
print("="*55)

print("\nHidden Layer Input")
print(np.round(hidden_input,4))

print("\nHidden Layer Output")
print(np.round(hidden_output,4))

print("\nPredicted Output")
print(np.round(prediction,4))

print("\nLoss")
print(np.round(loss,4))

print("\nGradient W1")
print(np.round(gradW1,6))

print("\nGradient W2")
print(np.round(gradW2,6))

print("\nUpdated W1")
print(np.round(new_W1,6))

print("\nUpdated W2")
print(np.round(new_W2,6))
```
